2. 핀터스텔라 라이브러리를 fs로 import

```python
import finterstellar as fs
```

3. 애플의 2020년 1월 1일 ~ 2020년 12월 31일 사이의 주가 데이터를 불러온다

```python
df = fs.get_price('AAPL', start_date='2020-01-01', end_date='2020-12-31')
```

4. 주가 데이터를 이용하여 주가 그래프 작성

```python
fs.draw_chart(df, right='AAPL')
```

5. 핀터스텔라 내장 함수 중 RSI를 측정하는 rsi함수를 이용하여 애플의 보조 지표를 작성
*   주가가 하락하는 2~3월에는 rsi 지수가 낮다.
*   주가가 상승하는 4~9월에는 rsi 지수가 높다.

```python
fs.rsi(df, w=14) # 14일 기준으로 RSI 측정
fs.draw_chart(df, left='rsi', right='AAPL')
```

6. RSI 40 매수 / 60 매도 전략

```python
fs.indicator_to_signal(df, factor='rsi', buy=40, sell=60)
fs.position(df)
fs.evaluate(df, cost=0.001)
fs.performance(df, rf_rate = 0.01)
```

7. RSI 40 매수 / 60매도 전략 대신 RSI 60 매수 / 40 매도 전략을 사용

```python
fs.indicator_to_signal(df, factor='rsi', buy=60, sell=40)
fs.position(df)
fs.evaluate(df, cost=0.001)
fs.performance(df, rf_rate = 0.01)
```
