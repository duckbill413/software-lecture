## 60/40 포트폴리오
- 한국 주식에 대해
- Portfolio Visualizer 백테스팅 결과중 복리수익률 그래프 출력 기능 python 코딩

```python
import FinanceDataReader as fdr
import pandas_datareader as pdr
import pandas as pd
from datetime import datetime, timedelta
import numpy as np
# pandas 표기 설정
pd.options.display.float_format = '{:.4f}'.format # 소수점 4자리까지 표기
pd.set_option('display.max_rows', None) # Dataframe 출력시 컬럼 생략없이 전체 표기
```

```python
# 종목의 시작일 부터 종가를 수집하는 함수
def get_price_data(stocks, start_data):
    df_price = pd.DataFrame(columns=stocks)
    for s in stocks:
        df_price[s] = fdr.DataReader(s, start=start_date)['Close']
    
    return df_price

stocks = ['KS11', 'KR10YT=RR'] # kospi와 10년만기 한국국채 수익률
start_date = '1998-01-01' # 시작일
dt = get_price_data(stocks, start_date)
dt
```

```python
# 월말 데이터 리샘플링
dt = dt.resample(rule='1Y').last()
dt['KR10YT=RR'] /= 100 # %단위 이전의 상태로 만들어준다.
dt
```

```python
#  KOSPI 월별 수익률, 누적 수익률, 월별 log 수익률, 누적 log 수익률 계산
dt['KS11_P'] = dt['KS11'].pct_change()
dt['KS11_PA'] = (1+dt['KS11_P']).cumprod()-1
dt['KS11_LP'] = np.log(dt['KS11_P']+1)
dt['KS11_LPA'] = dt['KS11_LP'].cumsum()

dt[['KS11_P','KS11_PA','KS11_LP','KS11_LPA']] *= 100
dt
```

```python
# KR10Y=RR은 10년만기 한국국채 수익률이므로 P로 칼럼 수정
dt = dt.rename(columns={'KR10YT=RR':'KR10YT_P'})
```

```python
#  KR10YT 누적 수익률, 월별 log 수익률, 누적 log 수익률 계산
dt['KR10YT_PA'] = (1+dt['KR10YT_P']).cumprod()-1
dt['KR10YT_LP'] = np.log(dt['KR10YT_P']+1)
dt['KR10YT_LPA'] = dt['KR10YT_LP'].cumsum()
dt[['KR10YT_P', 'KR10YT_PA', 'KR10YT_LP', 'KR10YT_LPA']] *= 100
dt
```

```python
# 복리수익률 계산 함수
def CAGR(first, last, periods):
    return ((last/first)**(1/periods)-1) * 100
```

```python
dt['KS11_CAGR'] = np.nan
dt['KR10YT_CAGR'] = np.nan

# 코스피의 복리 수익률 계산
KS11_first = dt['KS11'].iloc[0]
print("KS11_first: ", KS11_first)

for row in range(0, len(dt)):
    dt['KS11_CAGR'][row] = CAGR(KS11_first, dt['KS11'].iloc[row], row+1)
dt
```

```python
# 10년 만기 국채 복리 수익률 계산
KR10YT_first = dt['KR10YT_P'].iloc[2]/100+1
print("KR10YT_first: ", KR10YT_first)

for row in range(2, len(dt)):
    dt['KR10YT_CAGR'][row] = CAGR(KR10YT_first, dt['KR10YT_PA'].iloc[row]/100+1, row-1)
dt
```

#### 60-40 자산 배분
- 코스피지수 60%, 국채 40% 투자 복리 수익률 구하기

```python
dt['6040CAGR'] = dt['KS11_CAGR']*0.6+dt['KR10YT_CAGR']*0.4
dt
```
